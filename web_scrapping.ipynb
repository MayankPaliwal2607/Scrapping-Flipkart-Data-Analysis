## Scrapping Data From E-Commerce Website ( Flipkart[ Samsung Mobiles ] )

```python
# Importing Liabraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
# Create empty list to store the product,feature,price,rating
products = []
ram = []
inches = []
camera = []
battery = []
warranty = []
ratings = []
prices = []
```

```python
pages = list(range(1,31))       # range(1,31) 1 is first page of website and 30 is last page of website we scrap the data
for page in pages:
    # URL of the website you want to scrape 
    req = requests.get(f"https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=samsung+mobiles%7CMobiles&requestId=fbaf7eb2-aedf-480d-b2d9-d4825467bda8&page={page}").text
    soup = BeautifulSoup(req,'html.parser')
    
    # Scrapping product name
    product = soup.findAll('div',class_='_4rR01T')
    for i in range(len(product)):
        products.append(product[i].text)
        len(products)
        
    # Scrapping specification of product from website
    commonclass = soup.findAll('li',class_='rgWa7D')  # We observe the classnames for the different specificationare under one div.So we need to apply some methods to extract features.
    for i in range(len(commonclass)):
        p = commonclass[i].text
        
        if ('RAM' in p):
            ram.append(p)
        
        elif ('Display' in p):
            inches.append(p)
        elif ('Camera' in p):
            camera.append(p)
        elif('Battery' in p):
            battery.append(p)
        elif ('Warranty' in p):
            warranty.append(p)
        
        # Extracting prices of phones from website
        price = soup.findAll('div',class_='_30jeq3 _1_WHN1')
        for i in range(len(price)):
            prices.append(price[i].text)
            len(prices)
        # Extracting Rating of product from website
        rating = soup.findAll('div',class_='_3LWZlK')
        for i in range(len(rating)):
            ratings.append(rating[i].text)
            len(ratings)
```

```python
# Printing length of each data we scrap from website
print(len(products))
print(len(ram))
print(len(inches))
print(len(camera))
print(len(battery))
print(len(warranty))
print(len(prices))
print(len(ratings))
```

```python
# Finally merging all features to single dataframe
df = pd.DataFrame({'Product Name':products[522],'RAM':ram[522],'Display':inches[slice(522)],'Camera':camera[slice(522)],'Battery':battery[slice(522)],'Warranty':warranty[slice(522)],'Price':prices[slice(522)],'Rating':ratings[slice(522)]})
df.head(50)   # First top 50 rows of data
```

```python
# Converting file to .csv
df.to_csv('flipkart_mobiles.csv',index=False)   # Naming file to (flipkart_mobiles.csv)
```
